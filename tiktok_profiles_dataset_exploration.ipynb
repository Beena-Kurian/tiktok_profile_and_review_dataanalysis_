### TIKTOK PROFILES DATASET - EXPLORATION AND CLEANING

```python
import pandas as pd

# Load the dataset
df1 = pd.read_csv("tiktok_profiles_dataset.csv")

# Display basic information about the dataset
print("Dataset Info:")
print(df1.info())
```

```python
# Remove unnecessary columns
df1_cleaned = df1.drop(['nickname','biography', 'create_time', 'account_id', 'url', 'profile_pic_url', 'bio_link','top_videos'], axis=1)

# Display the first few rows of the cleaned DataFrame to confirm the changes
(df1_cleaned.head())
```

```python
# Check if the 'id' column is unique
id_unique = df1_cleaned['id'].is_unique

# Print the result
if id_unique:
    print("The 'id' column is unique.")
else:
    print("The 'id' column is not unique.")

# Move the 'id' column to the front
df1_final = df1_cleaned[['id'] + [col for col in df1_cleaned.columns if col != 'id']]

# Display the first few rows to confirm the changes
(df1_final.head())
```

```python
# Display summary statistics for numerical columns
print("\nSummary Statistics:")
(df1_final.describe())
```


#### To select the top 15% of profiles based on high values, you can follow these steps:

* Calculate the 85th percentile for each criterion (likes, followers, engagement rate).
* Filter the profiles that exceed these percentiles for each criterion.
* Calculate a composite score based on these filtered profiles.
* Sort the profiles based on the composite score.
* Select the top 15% of profiles based on this score.

```python
# Define weights for each criterion
weight_likes = 0.4
weight_followers = 0.4
weight_engagement = 0.2

# Calculate the 15th percentile of each criterion
likes_15th_percentile = df1_final['likes'].quantile(0.99)
followers_15th_percentile = df1_final['followers'].quantile(0.99)
engagement_15th_percentile = df1_final['awg_engagement_rate'].quantile(0.99)

# Calculate the composite score for each profile
df1_final.loc[:, 'composite_score'] = (weight_likes * (df1_final['likes'] >= likes_15th_percentile) +
                                       weight_followers * (df1_final['followers'] >= followers_15th_percentile) +
                                       weight_engagement * (df1_final['awg_engagement_rate'] >= engagement_15th_percentile)).values


# Sort the DataFrame by the composite score in descending order
df1_sorted = df1_final.sort_values(by='composite_score', ascending=False)

# Select the top 15% of profiles based on the composite score
total_profiles = len(df1_sorted)
top_15_percent_profiles = df1_sorted.head(int(0.15 * total_profiles))
```

```python
# Filter the DataFrame to get the top 15 accounts
top_15_accounts = df1_sorted.nlargest(15, 'composite_score')

# Count the number of verified accounts in the top 15
num_verified_accounts = top_15_accounts['is_verified'].sum()

print("Number of verified accounts among the top 15:", num_verified_accounts)
```

```python
# Calculate product of each feature
df1_sorted['product'] = df1_sorted['awg_engagement_rate'] * df1_sorted['comment_engagement_rate'] * df1_sorted['like_engagement_rate'] * df1_sorted['followers']

# # Define score range
min_score = df1_sorted['product'].min()
max_score = df1_sorted['product'].max()

# # Function to assign score based on value
def assign_score(value, min_val, max_val):
    score_range = max_val - min_val
    if score_range == 0:
        return 0
    normalized_value = (value - min_val) / score_range
    return normalized_value

# # Assign scores to each row based on product value
df1_sorted['score'] = df1_sorted['product'].apply(lambda x: assign_score(x, min_score, max_score))
min_score_range = df1_sorted['score'].min()
max_score_range = df1_sorted['score'].max()

print("Minimum score range:", min_score_range)
print("Maximum score range:", max_score_range)

min_score = df1_sorted['score'].min()
max_score = df1_sorted['score'].max()
df1_sorted['normalized_score'] = ((df1_sorted['score'] - min_score) / (max_score - min_score)) * 100

# Convert normalized scores to integers
df1_sorted['int_score'] = df1_sorted['normalized_score'].astype(int)
df1_sorted['int_score']
```

### STATISTICAL TEST-1

```python
from scipy import stats

# Separate data into verified and non-verified accounts
verified_engagement = df1_sorted[df1_sorted['is_verified']]['awg_engagement_rate']
non_verified_engagement = df1_sorted[~df1_sorted['is_verified']]['awg_engagement_rate']

# Perform two-sample t-test
t_statistic, p_value = stats.ttest_ind(verified_engagement, non_verified_engagement, equal_var=False)

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Check significance
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in engagement rates.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in engagement rates.")
```

* T-statistic: The t-statistic measures the difference between the means of the two groups relative to the variation in the data. In this case, the negative value suggests that the mean engagement rate of verified accounts is lower than that of non-verified accounts.

* P-value: The p-value is a measure of the probability of observing the data given that the null hypothesis is true. A p-value below the chosen significance level (alpha = 0.05) indicates that the observed difference in engagement rates is unlikely to occur by random chance alone. In this case, the p-value is very small (2.6946605001341477e-06), well below the significance level, suggesting strong evidence to reject the null hypothesis.

* Conclusion: Based on the low p-value, we reject the null hypothesis, concluding that there is a significant difference in engagement rates between verified and non-verified accounts on TikTok.

* The negative T-statistic and the rejection of the null hypothesis indicate that, on average, verified accounts tend to have lower engagement rates compared to non-verified accounts. In other words, the category of non-verified accounts tends to have higher engagement rates than verified accounts.

```python
from scipy.stats import mannwhitneyu

# Perform Mann-Whitney U Test
statistic, p_value = mannwhitneyu(verified_engagement, non_verified_engagement)

# Print the results
print("Mann-Whitney U Test Statistic:", statistic)
print("P-value:", p_value)

# Check significance
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in engagement rates.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in engagement rates.")
```
